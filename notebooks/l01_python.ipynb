#### 基本型

```python
# 整数型
# 数値表現が整数の場合、代入先変数は自動的に整数型になります。
a = 1

# 浮動小数点数型
# 数値表現に小数点が含まれると、代入先変数は自動的に浮動小数点型になります。
b = 2.0

# 文字列型
# 文字列はシングルクオート(')で囲みます。
# あるいはダブルクオート(")でもいいです。
c = 'abc'

# ブーリアン型
# True または False と取る変数の型です。
d = True
```

#### print関数とtype関数

```python
# 整数型変数aの値とtype
print(a)
print(type(a))
```

```python
# 浮動小数点数型変数bの値とtype
print(b)
print(type(b))
```

```python
# 文字列型変数cの値とtype'
print(c)
print(type(c))
```

```python
# ブーリアン型変数dの値とtype'
print(d)
print(type(d))
```

#### リスト

```python
# リストの定義
l = [1, 2, 3, 5, 8, 13]

# リストの値とtype
print(l)
print(type(l))
```

#### リストの要素数

```python
# リストの要素数
print(len(l))
```

#### リストの要素参照

```python
# リストの要素参照

# 最初の要素
print(l[0])

# 3番目の要素
print(l[2])

# 最後の要素 (こういう指定方法も可能)
print(l[-1])
```

#### 部分リスト参照1

```python
# 部分リスト インデックス:2以上 インデックス: 5未満
print(l[2:5])

# 部分リスト インデックス:0以上 インデックス: 3未満
print(l[0:3])

# 開始インデックスが0の場合は省略可
print(l[:3])
```

#### 部分リスト参照2

```python
# 部分リスト インデックス:4以上最後まで
# リストの長さを求める
n = len(l) 
print(l[4:n])

# 最終インデックスが最終要素の場合は省略可
print(l[4:])

# 後ろから2つ
print(l[-2:])

#最初も最後も省略するとリスト全体になる
print(l[:])
```

### 辞書

#### 辞書の定義

```python
# 辞書の定義
my_dict = {'yes': 1, 'no': 0}

# print文の結果
print(my_dict)

# type関数の結果
print(type(my_dict))
```

#### 辞書の参照

```python
# キーから値を参照

# key= 'yes'で検索
value1 = my_dict['yes']
print(value1)

# key='no'で検索
value2 = my_dict['no']
print(value2)
```

#### 辞書への項目追加

```python
# 辞書への項目追加
my_dict['neutral'] = 2

# 結果確認
print(my_dict)
```

### 制御構造

#### ループ処理

```python
# ループ処理

# リストの定義
list4 = ['One', 'Two', 'Three', 'Four']

# ループ処理
for item in list4:
    print(item)
```

```python
# range関数を使ったループ処理

for item in range(4):
    print(item)
```

```python
# 引数2つのrange関数

for item in range(1, 5):
    print(item)
```

```python
# 辞書とループ処理

# items関数
print(my_dict.items())

# items関数を使ったループ処理

for key, value in my_dict.items():
    print(key, ':', value )
```

#### if文

```python
# if文のサンプル
for i in range(1, 5):
    if i % 2 == 0:
        print(i, 'は偶数です')
    else:
        print(i, 'は奇数です')
```

####  関数

```python
# 関数の定義例
def square(x):
    p2 = x * x
    return p2

# 関数の呼出し例
x1 = 13
r1 = square(x1)
print(x1, r1)
```

```python
#  関数の定義例2
def squares(x):
    p2 = x * x
    p3= x * x * x
    return (p2, p3)

# 関数の呼出し例
x1 = 13
p2, p3 = squares(x1)
print(x1, p2, p3)
```

#### 数値の整形表示

```python
# f文字列の表示
a1 = 1.0/7.0
a2 = 123

str1 = f'a1 = {a1}   a2 = {a2}'
print(str1)
```

```python
# f文字列の詳細オプション

# .4f 小数点以下4桁の固定小数点表示
# 04 整数を0詰め4桁表示
str2 = f'a1 = {a1:.4f}  a2 = {a2:04}'
print(str2)

# 04e 小数点以下4桁の浮動小数点表示
# #x 整数の16進数表示
str3 = f'a1 = {a1:.04e}  a2 = {a2:#x}'
print(str3)
```
